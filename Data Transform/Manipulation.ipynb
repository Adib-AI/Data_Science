```python
import pandas as pd
import os
```

```python
dataset = pd.read_excel("Data - Survey Monkey Edited.xlsx", sheet_name = "Edited_Data")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset_modify = dataset.copy() #Menyimpan copy-an data asli untuk dimodifikasi
```

```python
dataset_modify.head(5)
```

```python
dataset_modify.columns
```

```python
column_to_drop = ['Start Date', 'End Date', 'Email Address',
       'First Name', 'Last Name', 'Custom Data 1']
column_to_drop
```

```python
dataset_modify = dataset_modify.drop(column_to_drop, axis = 1)
```

```python
dataset_modify.head(5)
```

```python
id_vars = list(dataset_modify.columns[:8]) #Mengambil nama kolom sampai 0-7 dan dimasukkan kedalam list
id_vars
```

```python
value_vars = list(dataset_modify.columns[8:])
value_vars
len(value_vars) #Data di tranposekan menjadi row sebuah column dan diiringi dengan value yang 
#terdapat pada dataset
```

```python
#Melth Method
dataset_melth = dataset_modify.melt(id_vars = id_vars, value_vars = value_vars, var_name = 'Question + Sub Question', value_name='Answer')
dataset_melth #default adalah variabel and value apabila tidak terdapat var_name and value_name
```

```python
question = pd.read_excel("Data - Survey Monkey Edited.xlsx", sheet_name = 'Question')
question.head(5)
```

```python
question2 = question.copy()
```

```python
question2.columns
```

```python
column_to_drop = ['Raw Question', 'Raw SubQuestion', 'Subquestion']
```

```python
question2 = question2.drop(column_to_drop, axis=1)
```

```python
question2.dropna(inplace = True) #Menghapus data yang "NaN"
```

```python
question2.head(5)
```

```python
#Merge Method
dataset_question2_merge = pd.merge(left=dataset_melth, right=question2, how="left", left_on="Question + Sub Question", 
                                   right_on="Question + Sub Question")
#Left = dataset mana yang diletakkan di kiri
#right = dataset mana yang diletkkan di kanan
#how = left,right,inner join (Sistem SQL, which mean "left" adalah left join)
#left_on = column mana yang di letkkan di kiri
#right_on = column mana yang di letkkan di kanan 
```

```python
print("Original Data", len(dataset_melth))
print("Merged Data", len(dataset_question2_merge))
```

```python
dataset_question2_merge
```

```python
respondend = dataset_question2_merge.copy()
```

```python
respondend = dataset_question2_merge[dataset_question2_merge["Answer"].notna()]
#Not Na adalah mengambil data yang tidak NaN
```

```python
respondend = respondend.groupby(['Question'])["Respondent ID"].nunique().reset_index() #Di groupby kan adalah Question, dan Respondent ID yang akan 
#aggeragate(nunique (mencari data yang memiliki ciri khas-nya sendiri)) kan adalah Answer. reset_index untuk menjadikan dataframe
respondend
```

```python
respondend.rename(columns = {"Respondent ID" : "Respondents"}, inplace = True)
respondend
```

```python
dataset_same_answer_merge = pd.merge(left=dataset_question2_merge, right=respondend, how="left",
                                     left_on="Question", right_on="Question")

print("Original Data", len(dataset_question2_merge))
print("Merged Data", len(dataset_same_answer_merge))
dataset_same_answer_merge
```

```python
same_answer = dataset_question2_merge
same_answer = same_answer.groupby(["Question + Sub Question","Answer"])["Respondent ID"].nunique().reset_index()
same_answer.rename(columns={"Respondent ID" : "Same Answer"}, inplace = True)
same_answer
```

```python
dataset_final_merge = pd.merge(left=dataset_same_answer_merge, right=same_answer, how="left", 
                               left_on=["Question + Sub Question", "Answer"], 
                               right_on=["Question + Sub Question", "Answer"])

dataset_final_merge["Same Answer"].fillna(0,inplace=True) #Mengganti "NaN" menjadi 0
print("Original Data", len(dataset_same_answer_merge))
print("Merged Data", len(dataset_final_merge))
dataset_final_merge
```

```python
dataset_final_merge.columns
```

```python
dataset_final_merge.rename(columns = {'Identify which division you work in.-Response':'Division Primary',
                                   'Identify which division you work in.-Other (please specify)':'Division Secondary',
                                   'Which of the following best describes your position level?-Response':'Position',
                                   'Which generation are you apart of?-Response':'Generation',
                                   'Please select the gender in which you identify.-Response':'Gender',
                                   'Which duration range best aligns with your tenure at your company?-Response':'Tenure',
                                   'Which of the following best describes your employment type?-Response':'Employement Type',
                                  }, inplace = True)
```
