```python
#Melakukan korelasi
korelasi = dataset.corr()
korelasi
```

```python
#Visualisasi dari korelasi
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.heatmap(korelasi)
```

```python
colom_korelasi = ['Pregnancies', 'Glucose', 'BMI', 'Age']
```

```python
def plot_boxplot(kolom, data):
    fig, ax = plt.subplots(1, 1, figsize=(5,5))
    sns.boxplot(x = kolom, data = data)
```

```python
for x in colom_korelasi:
    plot_boxplot(x, dataset)
```

```python
#Menggunakan Tukey IQR untuk mengatasai Outlier
def tentukan_outlier(x):
    q1,q3 = np.percentil(x, [25,75])
    
    #Buat IQR (q3-q1)
    iqr = q3-q1
    
    #Buat batasan bawah (floor) dan atas (ceiling)
    floor = q1 - (1.5*iqr)
    ceiling = q3 + (1.5*iqr)
    
    #Buat aturan dari floor dan ceiling untuk nilai mana yang termasuk outlier
    indikasi_nilai_outlier = list(x.index[(x < floor)| (x > ceiling)])
    nilai_outlier = list(x[indikasi_nilai_outlier])
    
    return indikasi_nilai_outlier, nilai_outlier
```

```python
for i in colom_korelasi:
    indikasi_nilai_outlier, nilai_outlier = tentukan_outlier(dataset[i])
    print("")
```
